```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
# Load the image
image_path = "eiffel1.jpg"
img = cv2.imread(image_path) # By default it reads in BGR order (not in RGB order)
```

```python
plt.imshow(img)
```

```python
# If you wanna interchange into RGB order from BGR, this can be done.
img_rgb = np.dstack([img[:,:,2], img[:,:,1], img[:,:,0]])
plt.imshow(img_rgb)
```

```python
# Now, lets convert into grayscale.
img_gray = img[:,:,0]/3 + img[:,:,1]/3 + img[:,:,2]/3
print(img_gray.shape)
plt.imshow(img_gray)
```

```python
# In the above image, image is converted to grayscale, but it's looking weired with blue yellow green colors.
# It's because Matplotlib is adding the colors, thats why we use cmap.
plt.imshow(img_gray, cmap = "gray")
```
