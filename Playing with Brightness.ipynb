```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
# Load the image
image_path = "eiffel1.jpg"
img = cv2.imread(image_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
# Show the image
plt.imshow(img)
```

```python
# Printing the shape of the image
print(img.shape)
```

```python
# Printing the img array
img
```

```python
# Show the Red Layer of the RGB Image
plt.imshow(img[:,:,0], cmap = "Reds")
```

```python
# Things to learn
# As the pixel value exist from 0 to 255 
# Suppose a pixel is at value 130, and you increase by 40 (towards the white), pixel value will now be 170.
# But, if a pixel value is 240, and you increase by 40, it will not be 280, it will be 24, 
# as it'll go till 255 and then start from 0. So, I'm limiting the pixel values to 255.

print(img[:,:,0])
print("---------------------")
print(img[:,:,0] + 40)
```

```python
# Now generating the each channel of RGB, and using numpy dstack.
level = 30
rgb_new = np.dstack([np.where(img[:,:,i] < (255 - level), img[:,:,i] + level, 255) for i in range(3)])
plt.figure(figsize = (25,6))
plt.subplot(1,3,1)
plt.imshow(img)
plt.subplot(1,3,2)
plt.imshow(rgb_new)
```

```python
# You can clearly see the difference of white level in both the images.
```
